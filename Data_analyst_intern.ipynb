**all queries from assignment **

```python
# 1) Create a cohort based on month of first session

query = """
WITH UserCohorts AS (
    SELECT
        user_id,
        strftime('%Y-%m', MIN(session_date)) AS cohort_month
    FROM sessions
    GROUP BY user_id
)
SELECT
    cohort_month,
    COUNT(user_id) AS total_users
FROM UserCohorts
GROUP BY cohort_month
ORDER BY cohort_month;
"""

result = pd.read_sql_query(query, conne)
result
```

```python
# Calculate retention (%) for session 2, 3, 4
query = """
SELECT
    session_number,
    COUNT(DISTINCT user_id) AS users,
    ROUND(
        COUNT(DISTINCT user_id) * 100.0 /
        (SELECT COUNT(DISTINCT user_id)
         FROM sessions
         WHERE session_number = 1),
    2) AS retention_percentage
FROM sessions
WHERE session_number IN (1,2,3,4)
GROUP BY session_number
ORDER BY session_number;
"""

df = pd.read_sql_query(query, conne)
df
```

```python
# Q.3) Total users → users who completed session 1 → session 2 → session 3 and show conversion % at each step

query = """
WITH total_users AS (
    SELECT COUNT(DISTINCT user_id) AS users FROM users),
session1 AS (
    SELECT COUNT(DISTINCT user_id) AS users FROM sessions WHERE session_number=1),
session2 AS (
    SELECT COUNT(DISTINCT user_id) AS users FROM sessions WHERE session_number=2),
session3 AS (
    SELECT COUNT(DISTINCT user_id) AS users FROM sessions WHERE session_number=3)


SELECT 'Total Users' AS stage, users, 100.0 AS conversion_percentage FROM total_users

UNION ALL

SELECT 'Session 1', users,
ROUND(users * 100.0 / (SELECT users FROM total_users), 2)
FROM session1

UNION ALL

SELECT 'Session 2', users,
ROUND(users * 100.0 / (SELECT users FROM session1), 2)
FROM session2

UNION ALL

SELECT 'Session 3', users,
ROUND(users * 100.0 / (SELECT users FROM session2), 2)
FROM session3;

"""

pd.read_sql(query, conne)
```

```python
# Calculate:

# Total revenue
query = """
SELECT
    SUM(fee) AS total_revenue
FROM sessions;
"""

pd.read_sql(query, conne)
```

```python
# Revenue per user

query = """
SELECT
    user_id,
    SUM(fee) AS revenue_per_user
FROM sessions
GROUP BY user_id;
"""

df= pd.read_sql(query, conne)
df.head(20)
```

```python
# Revenue per cohort
query = """
WITH cohort AS (
    SELECT
        user_id,
        strftime('%Y-%m', MIN(session_date)) AS cohort_month
    FROM sessions
    GROUP BY user_id
)
SELECT
    c.cohort_month,
    SUM(s.fee) AS total_revenue
FROM sessions s
JOIN cohort c ON s.user_id = c.user_id
GROUP BY c.cohort_month
ORDER BY c.cohort_month;

"""

pd.read_sql(query, conne)
```

```python
# Revenue per therapist
query = """
SELECT
    therapist_id,
    SUM(fee) AS total_revenue
FROM sessions
GROUP BY therapist_id
ORDER BY total_revenue DESC;
"""

pd.read_sql(query, conne)
```

```python

# If a therapist completed more than 10 sessions in a month we are providing 20% of the total revenue as the bonus calculate monthly bonus for them.

query = """
SELECT
    therapist_id,
    strftime('%Y-%m', session_date) AS month,
    COUNT(*) AS total_sessions,
    SUM(fee) AS total_revenue,

    CASE
        WHEN COUNT(*) > 10
        THEN SUM(fee) * 0.2
        ELSE 0
    END AS bonus

FROM sessions
GROUP BY therapist_id, month
order by total_sessions desc;
"""

result = pd.read_sql_query(query, conne)
result
```

```python
# Identify at which session most users drop off and provide % distribution
query = """
SELECT
    session_number,
    COUNT(DISTINCT user_id) AS users
FROM sessions
GROUP BY session_number
ORDER BY session_number;
"""

df_sessions = pd.read_sql_query(query, conne)


df_sessions['drop_off_users'] = df_sessions['users'].shift(1) - df_sessions['users']


df_sessions['drop_off_%'] = (
    df_sessions['drop_off_users'] / df_sessions['users'].shift(1) * 100
).round(2)

df_sessions
```

```python
# Filling rating data with taking average
df['rating'] = df['rating'].fillna(df['rating'].mean())
df.head(11)
```

```python
df = df.sort_values(['user_id', 'session_date'])
df['days_between_sessions'] = df.groupby('user_id')['session_date'].diff().dt.days
df['days_between_sessions'] = df['days_between_sessions'].fillna(0)
df.head(11)
```

```python
df['review_text'] = df['review_text'].fillna('No Review')
df.head(21)
```

```python
# Validate session_number sequence
sessions = sessions.sort_values(['user_id','session_number'])

sessions['valid_sequence'] = sessions.groupby('user_id')['session_number'].diff().fillna(1)

invalid = sessions[sessions['valid_sequence'] != 1]
sessions.head(20)
```

```python
# Average rating per therapist
avg_rating_therapist = df.groupby('therapist_id')['rating'].mean()

avg_rating_therapist
```

```python
# Correlation between rating and retention
df = sessions.merge(feedback, on='session_id')

df['retained'] = df['session_number'] > 1

df[['rating','retained']].corr()
```

```python
# create user_lifetime_sessions

df['user_lifetime_sessions'] = df.groupby('user_id')['session_id'].transform('count')
```

```python
# create avg_rating

df['avg_rating'] = df.groupby('user_id')['rating'].transform('mean')
```

```python
# create days_between_sessions

df = df.sort_values(['user_id', 'session_date'])
df['session_date'] = pd.to_datetime(df['session_date'])
df['days_between_sessions'] = df.groupby('user_id')['session_date'].diff().dt.days
```
